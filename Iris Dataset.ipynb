```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = pd.read_csv("iris.csv")
```

```python
iris.head()
```

```python
# ScatterPlots

# Here we are plotting sepal_length vs sepal_width 
# setosa - 'red'; versicolor - 'blue'; virginica - 'green'

for n in range(0,150):
    if iris['Name'][n] == 'Iris-setosa':
        plt.scatter(iris['SepalLength'][n], iris['SepalWidth'][n], color = 'red')
        plt.xlabel('SepalLength')
        plt.ylabel('SepalWidth')
    elif iris['Name'][n] == 'Iris-versicolor':
        plt.scatter(iris['SepalLength'][n], iris['SepalWidth'][n], color = 'blue')
        plt.xlabel('SepalLength')
        plt.ylabel('SepalWidth')
    elif iris['Name'][n] == 'Iris-virginica':
        plt.scatter(iris['SepalLength'][n], iris['SepalWidth'][n], color = 'green')
        plt.xlabel('SepalLength')
        plt.ylabel('SepalWidth')
```

**Pair Plots**

If we have "n" number of features in a data, Pair plot will help us create a ("n x n") figure where the diagonal plots will be histogram plot of the feature corresponding to that row and rest of the plots are the combination of feature from each row in y axis and feature from each column in x axis.

```python
# Pair Plots

sns.set_style("whitegrid");
sns.pairplot(iris, hue="Name", size=3);
plt.show()
```

**Box Plots**

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution.

```python
# Box Plots
# Plotting the features using boxes

plt.subplots(figsize=(12,10))
plt.style.use('ggplot')
plt.subplot(2,2,1)
sns.boxplot(x = 'Name', y = 'SepalLength', data = iris)
plt.subplot(2,2,2)
sns.boxplot(x = 'Name', y = 'SepalWidth', data = iris)
plt.subplot(2,2,3)
sns.boxplot(x = 'Name', y = 'PetalLength', data = iris)
plt.subplot(2,2,4)
sns.boxplot(x = 'Name', y = 'PetalWidth', data = iris)
```

**Violin Plots**

The violin plots (from seaborn package) can be inferred as a combination of Box plot at the middle and distribution plots (Kernel Density Estimation ) on both side of the data. This can give us the details of distribution like whether the distribution is mutimodal, Skewness etc.

```python
# Violin Plots
# Representing data using violin form

plt.subplots(figsize=(12,10))
plt.style.use('ggplot')
plt.subplot(2,2,1)
sns.violinplot(x = 'Name', y = 'SepalLength', data = iris)
plt.subplot(2,2,2)
sns.violinplot(x = 'Name', y = 'SepalWidth', data = iris)
plt.subplot(2,2,3)
sns.violinplot(x = 'Name', y = 'PetalLength', data = iris)
plt.subplot(2,2,4)
sns.violinplot(x = 'Name', y = 'PetalWidth', data = iris)
```

**Joint Plots**

Join plots can do both univariate as well as bivariate analysis. The main plot will give us a bivariate analysis, whereas on the top and right side we will get univariate plots of both the variables that were considered. It makes our job easy by getting both scatter plots for bivariate and Distribution plot for univariate, both in a single plot.

```python
# Joint plots shows bivariate scatterplots
# And univariate histograms

sns.jointplot(x = 'SepalLength', y = 'SepalWidth', data = iris)
```

**Strip Plots**

A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.
It is is a graphical data anlysis technique for summarizing a univariate data set. It is typically used for small data sets.

```python
# Strip Plots
# Plottign data in strip

plt.subplots(figsize=(12,10))
plt.subplot(2,2,1)
sns.stripplot(x = 'Name', y = 'SepalLength', data = iris, jitter = True)
plt.subplot(2,2,2)
sns.stripplot(x = 'Name', y = 'SepalWidth', data = iris, jitter = True)
plt.subplot(2,2,3)
sns.stripplot(x = 'Name', y = 'PetalLength', data = iris, jitter = True)
plt.subplot(2,2,4)
sns.stripplot(x = 'Name', y = 'PetalWidth', data = iris, jitter = True)
```

**Lmplot() Function**

Seaborns lmplot is a 2D scatterplot with an optional overlaid regression line. Logistic regression for binary classification is also supported with lmplot . It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.

The fuction can draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line with a 95% confidence interval for that regression.

lmplot() has data as a required parameter and the x and y variables must be specified as strings.

```python
# This graph is same as above but plotting the species separately

sns.lmplot(x = 'SepalLength', y = 'SepalWidth', data = iris, hue = 'Name', col = 'Name')
```
